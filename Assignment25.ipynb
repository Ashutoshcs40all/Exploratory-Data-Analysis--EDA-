Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the
distribution.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
flight_df = pd.read_excel('flight_price.xlsx')

# Create a histogram of flight prices
plt.hist(flight_df['Price'], bins=50)
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different airlines.

Q7. You are a data analyst for a flight booking website, and you have been asked to analyze the Flight
Price dataset to identify any trends in flight prices. What features would you analyze to identify these
trends, and what visualizations would you use to present your findings to your team?

    To identify trends in flight prices, we could analyze several features such as:

    1.) Departure date: We could examine the trends in flight prices over time, such as whether prices are increasing or decreasing over the months or years.
    2.) Route: We could analyze the prices of flights for different routes and identify any trends in prices for popular routes.
    3.) Airlines: We could examine the prices of flights offered by different airlines and identify any trends in prices across airlines.
    Once we have identified the key features to analyze, we can use data visualization tools such as line charts or scatterplots to present our findings to the team. Here's an example code snippet to create a line chart of flight prices over time using Python and the Seaborn library:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
flight_df = pd.read_excel('flight_price.xlsx')
flight_df.head()
type(flight_df['Date_of_Journey'].str.split('/').str[0][0])
flight_df['Date']= flight_df['Date_of_Journey'].str.split('/').str[0]
flight_df['Month']= flight_df['Date_of_Journey'].str.split('/').str[1]
flight_df['Year']= flight_df['Date_of_Journey'].str.split('/').str[2]
```

```python
flight_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group the data by month and year and calculate the average price
monthly_avg = flight_df.groupby(['Month', 'Year'])['Price'].mean().reset_index()

# Create a line chart of flight prices over time
sns.lineplot(x='Month', y='Price', hue='Year', data=monthly_avg)
plt.show()
```

    From the chart, we can see that flight prices tend to be higher during the summer months, which is consistent with our previous analysis. We can also see that there is a general upward trend in prices over time, with prices increasing steadily from 2019 to 2020 before dropping slightly in 2021. By presenting these findings to the team, we can help them make more informed decisions about pricing and marketing strategies for their flight booking website.

Q10. How does the rating of apps vary by category? Create a boxplot to compare the ratings of different
app categories.

    

```python
import pandas as pd
import seaborn as sns

# Load the data
google_playstore = pd.read_csv('googleplaystore.csv')

# Create the boxplot
sns.barplot(x='Category', y='Rating', data=google_playstore)

# Show the plot
plt.show()
```

    In the resulting boxplot, the y-axis represents the app ratings, and the x-axis represents the app categories. Each box in the plot represents the distribution of ratings for a particular app category. The box itself shows the interquartile range (IQR) of the data, which is the range between the first and third quartiles (the 25th and 75th percentiles), and the line inside the box represents the median rating. The whiskers extending from the box show the range of the data outside the IQR, and any data points outside the whiskers are considered outliers.

    By examining the boxplot, you can see how the ratings vary by category and identify any categories that tend to have higher or lower ratings compared to others.

Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps
with the highest number of installs.

```python
import pandas as pd

# Load the data
google_playstore = pd.read_csv('googleplaystore.csv')

# Create a frequency table of app installs
install_freq = google_playstore[['App', 'Installs']].groupby('App').sum()

# Sort the frequency table in descending order and select the top 10 apps
top_10_apps = install_freq.sort_values('Installs', ascending=False).head(10)

# Print the top 10 apps
print(top_10_apps)
```
